```python
import pandas as pd
import matplotlib
import mplfinance as mpf
df = pd.read_csv('../newdata/DowJones_history.csv',index_col='Date',parse_dates=True )
for i in df.groupby('Symbol'):
    mc = mpf.make_marketcolors(up='g',down='r',inherit=True)
    s  = mpf.make_mpf_style(base_mpf_style='yahoo',marketcolors=mc)
    kwargs = dict(type='candle',tight_layout=True,mav=(5,20), volume=True, figratio=(12,8), figscale=1,title=i[0], style=s)
    mpf.plot(i[1].tail(60), **kwargs)
```
